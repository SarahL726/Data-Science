```python
!pip install folium
import folium
from folium.map import *
import requests
import pandas

arrest_table = pandas.read_csv("https://cmsc320.github.io/files/BPD_Arrests.csv")

arrest_table = arrest_table[pandas.notnull(arrest_table["Location 1"])]

arrest_table["lat"], arrest_table["long"] = arrest_table["Location 1"].str.split(",").str
arrest_table["lat"] = arrest_table["lat"].str.replace("(", "").astype(float)
arrest_table["long"] = arrest_table["long"].str.replace(")", "").astype(float)
```

```python
arrest_table.head()
```

```python
map_osm = folium.Map(location=[39.29, -76.61], zoom_start=11)
map_osm
```

```python
List = []
count = 0

# sex = FeatureGroup(name='Sex')
race = FeatureGroup(name='Race')

import random
while count < 500:
  idx = random.randint(0, len(arrest_table['race']))
  if idx not in List:
    count += 1
    List.append(idx)
```

```python
for i in List:
  if arrest_table['race'].values[i] == 'B':
    race.add_child(folium.Circle(
        radius = 10,
        location = [arrest_table['lat'].values[i], arrest_table['long'].values[i]],
        popup = "Race: " + str(arrest_table['race'].values[i]),
        color = 'black',
        fill = True,
    ))
  elif arrest_table['race'].values[i] == 'W':
    race.add_child(folium.Circle(
        radius = 10,
        location = [arrest_table['lat'].values[i], arrest_table['long'].values[i]],
        popup = "Race: " + str(arrest_table['race'].values[i]),
        color = 'white',
        fill = True,
    ))
  elif arrest_table['race'].values[i] == 'A':
    race.add_child(folium.Circle(
        radius = 10,
        location = [arrest_table['lat'].values[i], arrest_table['long'].values[i]],
        popup = "Race: " + str(arrest_table['race'].values[i]),
        color = 'yellow',
        fill = True,
    ))
  elif arrest_table['race'].values[i] == 'U':
    race.add_child(folium.Circle(
        radius = 10,
        location = [arrest_table['lat'].values[i], arrest_table['long'].values[i]],
        popup = "Race: " + str(arrest_table['race'].values[i]),
        color = 'green',
        fill = True,
    ))
  elif arrest_table['race'].values[i] == 'H':
    race.add_child(folium.Circle(
        radius = 10,
        location = [arrest_table['lat'].values[i], arrest_table['long'].values[i]],
        popup = "Race: " + str(arrest_table['race'].values[i]),
        color = 'brown',
        fill = True,
    ))
  elif arrest_table['race'].values[i] == 'I':
    race.add_child(folium.Circle(
        radius = 10,
        location = [arrest_table['lat'].values[i], arrest_table['long'].values[i]],
        popup = "Race: " + str(arrest_table['race'].values[i]),
        color = 'orange',
        fill = True,
    ))
```

```python
map_osm.add_child(race)
map_osm.add_child(folium.map.LayerControl())
```

```python
import os
from IPython.core.display import HTML
map_osm.save(os.path.join('map_osm.html'))
```

Since the spots are randomly chosen from the dataset for the marker, it can be viewed as completely random and is able to show the distribution of the arrests. 

The number of black spots are significantly higher of any other color, and the white spots are listed as second. Convert the observation to plain text, it means that the arrests of black people is much more than the rest, while white people are the second most. You can barely see other colors. Green spots can also be noticed.

Based on the distribution of the markers, it can be noticed that most arrests are made in the middle of baltimore, and most of them are done near the state highways. It can be indicated that most of the arrests are done when the suspects are on their way to escape or in a car. Surprisingly, the arrests on the sideways is not higher than the ones on the Avenues. So either Avenues are not as safe as we thought, or the arrests are more likely to be done on more open space.
